```python
from datascience import *
import numpy as np
```

```python
movies_table = Table.read_table('movies.csv')
ratings_table = Table.read_table('ratings.csv')
```

```python
ratings_grouped = ratings_table.group('movieId', np.mean)
movie_recommendations = movies_table.join('movieId', ratings_grouped)
```

```python
movie_recommendations.show(5)
```
